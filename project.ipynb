# VARIABLE TRANSFORMATION

```python
df.loc[:,"intdaty_dv"]=df.loc[:,"intdaty_dv"].astype(int)
```

```python
df=pd.merge(df, CPI, how="inner", left_on="intdaty_dv", right_index=True)
df
```

```python
df["paygu_dv"]=df["paygu_dv"].astype(float)
df["paynu_dv"]=df["paynu_dv"].astype(float)
```

```python
df["paygu_dv"]=df["paygu_dv"]/(df["CPI"]/100)
df["paynu_dv"]=df["paynu_dv"]/(df["CPI"]/100)
```

```python
df["urban_dv"]=df["urban_dv"].replace({"1":1, "2":2}).astype(int)
df["urban_dv"].unique()
```

```python
df["age_dv"]=df["age_dv"].astype(int)
df["age_dv"].unique()
```

```python
df["age_dv2"]=df["age_dv"]**2
df["age_dv2"].unique()
```

```python
df["sex_dv"]=df["sex_dv"].replace({"1":1, "2":2}).astype(int)
df["sex_dv"].unique()
```

```python
df["bornuk_dv"]=df["bornuk_dv"].replace({"1":1, "2":2, " ":1}).astype(int)
df["bornuk_dv"].unique()
```

```python
df["hhsize_dv"]=df["hhsize_dv"].astype(int)
df["hhsize_dv"].unique()
```

```python
df["mstat_dv"]=df["mstat_dv"].astype(int)
df["mstat_dv"].unique()
```

```python
df["hiqual_dv"]=df["hiqual_dv"].replace({" ":0}).astype(int)
df["hiqual_dv"].unique()
```

```python
df["jbstat"]=df["jbstat"].astype(int)
df["jbstat"].unique()
```

```python
df["jbnssec8_dv"]=df["jbnssec8_dv"].astype(int)
df["jbnssec8_dv"].unique()
```

```python
df["jbmngr"]=df["jbmngr"].astype(int)
df["jbmngr"].unique()
```

```python
df["jbft_dv"]=df["jbft_dv"].astype(int)
df["jbft_dv"].unique()
```

```python
df["j2has"]=df["j2has"].replace({"1":1, "2":2}).astype(int)
df["j2has"].unique()
```

```python
df["ethn_dv"]=df["ethn_dv"].replace({" ":0}).astype(int)
df["ethn_dv"].unique()
```

```python
df["jbterm_dv"]=df["jbterm_dv"].replace({" ":0}).astype(int)
df["jbterm_dv"].unique()
```

```python
df["jbhrs"]=df["jbhrs"].astype(float)
df["jbhrs"].unique()
```

```python
df["jbot"]=df["jbot"].replace().astype(int)
df["jbot"].unique()
```

```python
df["jbotpd"]=df["jbotpd"].replace().astype(int)
df["jbotpd"].unique()
```

**calculate the hourly rate**

```python
df.loc[:,"tot_hours_month"]=4*(df["jbhrs"]+df["jbot"]+df["jbotpd"])
df.loc[:,"hourly_real_gross_salary"]=df.loc[:,"paygu_dv"]/df.loc[:,"tot_hours_month"]
df.loc[:,"hourly_real_net_salary"]=df.loc[:,"paynu_dv"]/df.loc[:,"tot_hours_month"]
df
```

**transform hourly rate into log hourly rate**

```python
df.loc[:,"ln_hourly_real_gross_salary"]=np.log(df.loc[:,"hourly_real_gross_salary"])
df.loc[:,"ln_hourly_real_net_salary"]=np.log(df.loc[:,"hourly_real_net_salary"])
df
```
